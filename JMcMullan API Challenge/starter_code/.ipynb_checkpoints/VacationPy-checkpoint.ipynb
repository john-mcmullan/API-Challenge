```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from pprint import pprint

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
csv = pd.read_csv("../output_data/citydata.csv")
csv
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Store latitude and longitude in locations
locations = csv[["Lat", "Lng"]]

Hum = csv["Humidity"]
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=Hum, 
                                 dissipating=True, max_intensity=150,
                                 point_radius=15)
fig.add_layer(heat_layer)

fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
Hotel_df = csv.loc[(csv['Max Temp'] >= 21.1111) & (csv['Max Temp'] <= 26.66666667) & (csv['Wind Speed'] <= 10) & (csv['Cloudiness'] == 0), :]
Hotel_df['Hotel Name'] = ''
Hotel_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
#target_coordinates = "-20.7377, 116.8463"
target_radius = 5000
target_type = "lodging"
target_search = "Hotel"


parameters = {
    'keyword': target_search,
    'radius': target_radius,
    'type': target_type,
    'key': g_key
}

url = f'https://maps.googleapis.com/maps/api/place/nearbysearch/json'

print('Finding Hotels')     
print('-----------------------------')
for index, row in Hotel_df.iterrows():
    
    #grab type
    target_lat = row['Lat']
    target_lng = row['Lng']
    place = row['City']
    
    #inserts into parameters
    parameters['location'] = f"{target_lat}, {target_lng}"
    
    response = requests.get(url, params=parameters).json()
    
    #grab results
    results = response['results']
    
    try:
        #extract information
        Hotel_df.loc[index, 'Hotel Name'] = results[0]["name"]
        print(f'Found a hotel in {place}')
        
    except (KeyError, IndexError):
        print(f'no hotel available in {place}')
        
print("-----------------------------")
print("Data Retrieval Complete")   
print("-----------------------------")
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in Hotel_df.iterrows()]
locate = Hotel_df[["Lat", "Lng"]]
marker_layer = gmaps.marker_layer(locate, info_box_content=hotel_info)
```

```python
# Add marker layer ontop of heat map
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Display figure
fig
```
